```python
#Simplified import block...
import math
import numpy as np
import matplotlib.pyplot as plt
import mie_data_generator_function as mdg
```

```python
# Set up data range
bead_index = 1.55 #fix it at polystyrene
n_radii = 1
n_factors = 100
n_factor = np.linspace(1,1.5, n_factors)
n_angles = 500
#radius = np.linspace(6e-6, 10e-6, n_radii)  #same as the present model for now
radius = 8e-6
scatI = np.zeros([n_factors,n_angles], dtype = 'float')
theta = np.zeros([n_angles], dtype='float')
```

```python
#make a Mie data array
for ii in range(n_factors):
    theta, scatI[ii,:] = mdg.generate_data(bead_index, radius, n_factor[ii], n_angles)
    
#theta, scatI = mdg.generate_data(bead_index, radius, n_angles)

#plot one for kicks
fig, ax = plt.subplots()
ax.plot(theta, np.log10(scatI[0,:]))
ax.plot(theta, np.log10(scatI[n_factors-1,:]))
plt.show()
```

```python
#now look at the array overall
#now let's make a nice color plot of this
plt.figure()
plt.pcolor(theta, n_factor, np.log10(scatI))
plt.xlabel('angle')
plt.ylabel('n_factor')
plt.title('Mie scattering as a fn of approximation')
plt.colorbar()
plt.show()
```
